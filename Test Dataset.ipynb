```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
g = ['Male', 'Female']
gender = np.random.choice(g, size=100, p=(0.489, 0.511))

age = np.random.randint(low=18, high=65, size=100)

m= ['private vehicle', 'public transport', 'cycle', 'walk']
mode = np.random.choice(m, size=100, p=(0.62, 0.09, 0.11, 0.18))

data = {'Gender':gender, 'Age':age, 'Mode of Transport':mode}
```

```python
df = pd.DataFrame(data, columns=['Gender', 'Age', 'Mode of Transport', 'Journey Distance', 'Journey Time'])
```

```python
pv = df[df['Mode of Transport'] == 'private vehicle']
```

```python
pt = df[df['Mode of Transport'] == 'public transport']
```

```python
cyc = df[df['Mode of Transport'] == 'cycle']
```

```python
w = df[df['Mode of Transport'] == 'walk']
```

```python
# Journey Distances for cycling
d_cyc = np.random.triangular(left=2, mode=13, right=14, size=len(cyc))
jd_cyc = np.round(d_cyc, 0)
```

```python
# Journey time for cycling
jt_cyc = (jd_cyc / 20) * 60
```

```python
# Journey Distances for walking
d_w = np.random.triangular(left=0, mode=5, right=6, size=len(w))
jd_w = np.round(d_w, 0)
```

```python
# Journey time for walking
jt_w = (jd_w / 5) * 60
```

```python
# Journey Distances for public transport
d_pt = np.random.normal(loc=13.7, scale=4, size=len(pt))
jd_pt = np.round(d_pt, 0)
```

```python
# Journey time for public transport
jt_pt = (jd_pt / 30) * 60
```

```python
# Journey Distances for private vehicles
d_pv = np.random.normal(loc=13.7, scale=4, size=len(pv))
jd_pv = np.round(d_pv, 0)
```

```python
# Journey time for private vehicles
jt_pv = np.round((jd_pv / 30) * 60, 0)
```

```python
df.loc[df['Mode of Transport'] == 'walk', 'Journey Distance'] = jd_w
df.loc[df['Mode of Transport'] == 'walk', 'Journey Time'] = jt_w

df.loc[df['Mode of Transport'] == 'cycle', 'Journey Distance'] = jd_cyc
df.loc[df['Mode of Transport'] == 'cycle', 'Journey Time'] = jt_cyc

df.loc[df['Mode of Transport'] == 'private vehicle', 'Journey Distance'] = jd_pv
df.loc[df['Mode of Transport'] == 'private vehicle', 'Journey Time'] = jt_pv

df.loc[df['Mode of Transport'] == 'public transport', 'Journey Distance'] = jd_pt
df.loc[df['Mode of Transport'] == 'public transport', 'Journey Time'] = jt_pt
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
df
```

```python
mode_count = [len(pv), len(pt), len(cyc), len(w)]

plt.pie(mode_count, labels=m, explode = (0.1, 0.1, 0.1, 0.1), shadow=True, autopct='%1.1f%%')
plt.title("Pie Chart of Modes of Transport")
plt.show()
```
